```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_CatBoost = pd.read_csv("Mosaic_Minerals_3D/CatBoost/predictions.csv")
df_CatBoost.head()
```

```python
df_CatBoost[(df_CatBoost['predictions'] == 1) & (df_CatBoost['1_probability'] >= 0.75)]
```

```python
df_select1 = df_CatBoost[(df_CatBoost['predictions'] == 1) & (df_CatBoost['1_probability'] >= 0.92)]
fig = px.scatter_3d(df_select1, x='X', y='Y', z='Z', color='1_probability', range_color=(0,1),
                   color_continuous_scale=px.colors.sequential.Cividis_r,
                   title='CatBoost - 1 probability (Only values greater than 0.92 are shown)')
fig.update_traces(marker_size = 1)
fig.show()
```
